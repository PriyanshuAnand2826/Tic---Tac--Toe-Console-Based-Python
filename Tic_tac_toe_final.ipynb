```python
def choice_var():
    choice=''
    while (choice!='X' and choice!='O'):
        choice=input ('Please choose any one X or O = ').upper()
    if(choice=='X'):
        return('X','O')
    else:
        return('O','X')
    
from IPython.display import clear_output

def display_board(board):
    clear_output()
    print(board[7],'  |  ',board[8],'  | ',board[9])
    print(board[4],'  |  ',board[5],'  | ',board[6])
    print(board[1],'  |  ',board[2],'  | ',board[3])    

import random


def choose_first():
    flip= random.randint(0,1)
    if(flip==0):
        return 'Player 1 '
    else:
        return 'Player 2'
    
    
def player_position(board):
    position = 0
    while position not in [1,2,3,4,5,6,7,8,9] or  not space_check(board,position):
        position = int(input('enter the position(1-9) : '))
    return position


def player_input(board,position,choice):
    board[position]=choice

def win_check(board,mark):
    if(board[1]==board[2]==board[3]==mark or 
      board[4]==board[5]==board[6]==mark or 
      board[7]==board[8]==board[9]==mark or
      board[1]==board[4]==board[7]==mark or 
      board[2]==board[5]==board[8]==mark or
      board[3]==board[6]==board[9]==mark or
      board[1]==board[5]==board[9]==mark or
      board[3]==board[5]==board[7]==mark):
        return True 
    else:
        return False
    
def space_check(board,position):
    return board[position]==' '

def full_board_check(board):
    for i in range (1,10):
        if (space_check(board,i)):
            return False
        
    return True 

import random




def replay():
    response=input('want to play again ?? Yes or No = ').lower()
    return response=='yes'







print('Welcome to the game Guyss !!')
while True :
    #now we take the input for player1 and player 2 
    player1,player2=choice_var()
    the_board=[' ']*10
    first_move=player1
    print(first_move , " will go first ")
    play_game=input('ready to play ?? y or n = ').lower()
    
    if (play_game=='y'):
        game_on=True
    else:
        game_on=False
        break
    
        
    while game_on:
        if (first_move=='X'):
            #show the board
            display_board(the_board)
            #ask for position 
            position=player_position(the_board)
            #place the marker in that position
            player_input(the_board,position,player1)
            #check if won or not 
            if win_check(the_board,player1):
                display_board(the_board)
                print('Player 1 has won the game ')
                game_on=False
            else:
                if full_board_check(the_board):
                        display_board(the_board)
                        print('Match Tie !!')
                        game_on=False
                else:
                    if(player1=='X'):
                        first_move=player2
                    else:
                        first_move=player1
        else:
             #show the board
            display_board(the_board)
            #ask for position 
            position=player_position(the_board)
            #place the marker in that position
            player_input(the_board,position,player2)
            #check if won or not 
            if win_check(the_board,player2):
                display_board(the_board)
                print('Player 2 has won the game ')
                game_on=False
            else:
                if full_board_check(the_board):
                        display_board(the_board)
                        print('Match Tie !!')
                        game_on=False
                else:
                    if (player1=='O'):
                        first_move=player2
                    else:
                         first_move=player1
                     
                    
     
                    
    
    if not replay():
        break
        
        

```
